## deepcopy- 
- it  will copy the data to the differnt loaction and there will not be any kind of dependency. 
- x=[1,2,3,4,5]       
- z=np.copy(x)
-both of them having diffent differnt memory address. 

if we will make the change to the x it eill not gonna make any changes in z because of the diifernt memory address 

```python
x= np.array([1,2,3,4,5,6])
```

```python
z=np.copy(x)
z
```

```python
t=x.copy()
t
```

```python
x[0]=100
x
```

```python
t
```

```python
z     # no changes 
```

##  shollow -copy 
- here bothe x and y have the same memory address, and there is complete depency on each other if will make the change in one the changes will reflect in another one for sure.
- x=y ( same memory allocation)
- hence if will make the change in x or y , it definalty will make the chages in y and x, this is called as shollow cpy

see example:

```python
x= np.array([1,2,3,4,5,6])
```

```python
y=x
```

```python
y
```

```python
y[0]=100
```

```python
y
```
