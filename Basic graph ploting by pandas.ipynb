## histogram
- this will show the value range and frequency(number of occurances of particular data set)
- how many data set present between this range 
- what isthe frequency of data set in particvular range
alpha - intnsity of color(contrast of the graph

```python
df.plot(kind="hist",figsize=(10,8), alpha=0.8)
```

#### lets understand this by ploting one column only 
- the x axis of this is not showing indexes, the x-axis is showing value range

```python
df["SepalLengthCm"].plot(kind="hist", color="y", alpha=1)
```

```python
df["SepalLengthCm"].plot(kind="hist",orientation="horizontal",color="b")
```

```python
df.hist(figsize=(15,8),color="r")     # CALLING HISTOGRAM AS A FUNCTION 
```

## scatterplot  - 2D graph will have to give x axis and y axis

```python
df.plot(kind='scatter',x="SepalLengthCm",y="PetalWidthCm")
#df.plot.scatter(x="SepalLengthCm","SepalLengthCm")
```

```python
df.plot(kind='scatter',x="SepalLengthCm",y="PetalLengthCm",color="g")
```

```python
df.plot(kind='scatter',x="SepalLengthCm",y="PetalLengthCm",c="SepalWidthCm",s=500)    
# c is represnting color 
# s= size of dot
```

```python
df.plot(kind='scatter',x="SepalLengthCm",y="PetalLengthCm",c="SepalWidthCm",s=df["PetalWidthCm"]*500) 
```

## hexbin plot
