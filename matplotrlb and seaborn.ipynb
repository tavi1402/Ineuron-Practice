## hue in scatterplot (group by colors) (legand in plotly)
and this add colors differntiation or this is color by operation 

```python
df1['smoker'].value_counts()
```

```python
sns.relplot(data=df1, x='total_bill',y='tip', hue='smoker')
```

## style in sns, ( group by differnt styling)
(this also perform the same group by operation witrh different styling not with color)
- there is no difference in terms of prentation in hua dstyle.

```python
sns.relplot(data=df1, x='total_bill',y='tip', style='smoker')
```

```python
sns.relplot(data=df1, x='total_bill',y='tip', style='size',)
```

## we can opertae group by oeration hue and style together.

```python
sns.relplot(data=df1, x='total_bill',y='tip', style='time',hue='smoker',)
```

### size ( again performing group by opetration.)

```python
sns.relplot(data=df1, x='total_bill',y='tip',size= 'size')
```

```python
sns.relplot(data=df1, x='total_bill',y='tip',kind='line')
```
