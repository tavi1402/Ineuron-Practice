````python
India_data = px.data.gapminder().query("country == 'India' ")
fig = px.bar(India_data, x='year', y='pop', color='lifeExp', labels={'pop': 'Population of India'}, height=400)
fig.show() 
````

````python
# let's use plotly_dark in our  bar chart 
India_data = px.data.gapminder().query("country == 'India' ")
fig = px.bar(India_data, x='year', y='pop', color='lifeExp',
             labels={'pop': 'Population of India'},
             height=400, template='plotly_dark')
fig.show()
````

````python
# seaborn
fig = px.bar(India_data, x='year', y='pop', color='lifeExp', labels={'pop': 'Population of India'},
             height=400, template='seaborn')
fig.show()
````

### Create a graph using area

````python
fig = px.area(country_data, x="year", y="pop", color="continent", line_group="country")
fig.show()
````

### Plot Life Expectency vs GDP Per Capita

````python
country_data1992 = country_data.query('year == 1992')

# create scatter plot 
fig = px.scatter(country_data1992, x='gdpPercap', y='lifeExp',color='continent')
fig.show()
````

### Create Interactive Bubble Charts

````python
px.scatter(country_data1992, x="gdpPercap", y="lifeExp", color="continent", size="pop", size_max=60)
````

### Represent Geographic Data as Animated Maps using 

### create a map using scatter_geo()

````python
map_fig= px.scatter_geo(country_data,
                       locations='iso_alpha',
                       projection='orthographic',
                       color='continent',
                       opacity=.9,
                       hover_name='country',
                       hover_data=["lifeExp","pop","year"]
                       )
map_fig.show()
````

### create a map using line_geo()

````python

fig = px.line_geo(country_data, locations='iso_alpha', color='continent', projection='orthographic',hover_name='country')
fig.show()
````

### create a map using choropleth

````python
fig = px.choropleth(country_data, locations='iso_alpha', color='lifeExp', hover_name='country', 
                    animation_frame='year', color_continuous_scale=px.colors.sequential.Plasma, projection='natural earth')
fig.show() 
````

### creating  a facet plot 

````python

fig = px.scatter(country_data, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=60, 
                hover_name='country', facet_col='continent')
fig.show() 
````
