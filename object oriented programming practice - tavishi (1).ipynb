### square cube and squareroot using oops

```python
import math 
class calculator:
    def __init__(self,number):
        self.number=number
        print(f"The give number is {self.number}" )
    def square_of_number(self):
        print(f"the square of number {self.number} is ",math.pow((self.number),2))
    
    def cube_of_number(self):
        print(f"the cube of number {self.number} is ",math.pow((self.number),3))
              
    def square_root_of_number(self):
        print(f"the square root of number {self.number} is ",math.sqrt(self.number))
        
number1=calculator(2)  
number1.cube_of_number()
number1.square_of_number()
number1.square_root_of_number()

        
        
number2=calculator(8)  
number2.cube_of_number()
number2.square_of_number()
number2.square_root_of_number()
        
        
```

### static method

```python
# static method 
from datetime import date
class employee:
    company="Google"
    def __init__(self,name,gender,position,city):
        self.name=name
        self.gender=gender
        self.position=position
        self.city= city
        print(f"The Information of the programmers {self.name} working at {self.company} is:",)
        
    @staticmethod
    def greetings():
        print(f"Hello, Good Morning, Welcome to the company your joining date is ",date.today())
        print("\n")
        
    def get_details(self):
        print(f"The Company Name of the employee is {self.company}")
        print(f"The Name of the employee is {self.name}")
        print(f"The Gender of the employee is {self.gender}")
        print(f"The Position of the employee is {self.position}")
        print(f"The City of the employee belongs is {self.city}")
        print("\n")
        
    
Mavin=employee("Mavin","male","marketing manager","pune")
Mavin.greetings()
Mavin.get_details()



Erik=employee("Erik","male","HOD","Banglore")
Erik.get_details()


    
```

### inherritance

```python
from datetime import date
#parent or base class
class employee:
    company="Google"
    def __init__(self,name,gender,position,city):
        self.name=name
        self.gender=gender
        self.position=position
        self.city= city
        print(f"The Information of the programmers {self.name} working at {self.company} is:",)
        
    @staticmethod
    def greetings():
        print(f"Hello,Welcome to the company your joining date is ",date.today())
        print("\n")
        
    def get_details(self):
        print(f"The Company Name of the employee is {self.company}")
        print(f"The Name of the employee is {self.name}")
        print(f"The Gender of the employee is {self.gender}")
        print(f"The Position of the employee is {self.position}")
        print(f"The City of the employee belongs is {self.city}")
        print("\n")

        
# child or derieved class        
class emplyeee_char(employee):   
    print(f"the characterstics of the employee is:")
    
    @staticmethod
    def greetings():
        print(f"Hello, Good Morning, Welcome to the company your joining date is ",date.today())
        print("\n")
        
 
mavin=employee("Mavin","male","marketing manager","pune")
mavin.greetings()
mavin.get_details()


ved=emplyeee_char("ved","male","pilot","jodhpur")
ved.greetings()
ved.get_details()

```

### @classmethod

```python
class Employee:
    company="camel"
    salary=200000
    location="delhi"
    
# suppose we want to change the location of the employee permanenty we have two ,methods to do this 
# method1

    def change_location(self,loc):
        self.__class__.location=loc

e=Employee()
print(e.location)
e.change_location("Banglore")
print(e.location)
    
    
```

```python
# method2
class Employee:
    company="camel"
    salary=200000
    location="delhi"
    
    @classmethod
    def change_location(cls,loc):
        cls.location=loc

e=Employee()
print(e.location)
e.change_location("Banglore")
print(e.location)
    
    
```

### @.getter (@property decorator ), @.setter

```python
class Employee:
    company="camel"
    salary=200000
    salary_bonus= 40000
    #Total_salary=230000

    @property       
    def total_salary(self):
        return self.salary+self.salary_bonus
    
    @total_salary.setter
    def total_salary(self,val):
        self.salary_bonus=val-self.salary
    
e= Employee()
print(e.total_salary)
e.total_salary= 250000
print(e.salary)
e.salary_bonus
    
    
    
```
