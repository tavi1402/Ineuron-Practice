## reshape()

````python
a= np.fromfunction(lambda i,j,z : i*j*z,(4,3,2), dtype= int)  # total number of element are 24  
a
````

````python
'''
For 2D
total number of element are 24  this can be 
(1*24) or(24*1)
(12*2) or(2*12)=24 
(8*3) or (3*8)= 24

which menas any element whose multiplication is equal to 24 can reshape the array

---> we can reshape the same given data  i.e 3D into 4D.
possible combination will be 
(4,8,1,3)
(3,1,8,4)  and so on 


----> similarly we can convert this into 5-D, 6-D,7-D and so on
(4,8,1,3,1) ---> 5D
(4,1,8,1,3,1)---> 6D

'''

a.reshape(3,8)    
````

````python
a.reshape(2,12)
````

````python
n=a.reshape(6,4)
````

````python
# now we canvert this into a dataframe.
pd.DataFrame(n)
````
