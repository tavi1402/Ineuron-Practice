```python
import pandas as pd 
import numpy as np 
import plotly as plx
import cufflinks as cf
cf.go_offline()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv")
df
```

```python
df=df.drop("Id",axis=1)
```

```python
df
```

```python
df.iplot()
```

## scatterplot

```python
df.iplot(x="SepalLengthCm",y="PetalLengthCm")
```

```python
df.iplot(x="SepalLengthCm",y="PetalLengthCm", mode="markers",size=6)
# markers - create dots for you
```

```python
df.iplot(kind="scatter",x="SepalLengthCm",y="PetalLengthCm", mode="markers",size=8)
```

```python
df.iplot(kind="scatter",mode="markers", size=7)
```

## bubble plot

```python
df.iplot(kind="bubble", x="SepalLengthCm",y="PetalLengthCm",size="SepalWidthCm")
```

```python
df.scatter_matrix()
# representation of histogram and scatter
```
