### que1 = try to find out a count of each and every word in a respective file return a list of tuple with word and its respective count
sample example - [('sudh', 6 ) , ('kumar',3)]

```python
import csv

def wordCount(*args):
    words = []
    for file in args:
        with open(file, 'r', encoding="utf8") as f:
            data = csv.reader(f)
            for row_data in data:
                words.append(row_data[0])
    word_count = []
    unique_words = list(set(words))
    unique_words.sort()

    for i in unique_words:
        word_count.append((i, words.count(i)))
    print("length of word count is{} ".format(len(word_count)))  
        
    return (word_count)
```

```python
f1 = wordCount('vocab.enron.txt')
#f1
```

### que2 = try to perform a reduce operation to get a count of all the word starting with same alphabet
sample examle = [(a,56) , (b,34),...........]

```python
import csv

def wordCount_reduce(*args):

    words = []
    
    for file in args:
        with open(file, 'r', encoding="utf8") as f:
            data = csv.reader(f)
            for row_data in data:
                if row_data[0][0] >= 'a' and row_data[0][0] <= 'z' or row_data[0][0] >= 'A' and row_data[0][0] <= 'Z':
                    words.append(row_data[0][0])
                    
    word_count = []
    unique_words = list(set(words))
    unique_words.sort()


    for i in unique_words:
        word_count.append((i, words.count(i)))
        
    print("Data record count is {} : ".format(len(word_count)))
        
    return word_count
```

```python
f1 = wordCount_reduce('vocab.enron.txt')
f1
```

### que3 = Try to filter out all the words from dataset . .001.abstract = abstract =.002 = delete

```python
import csv

def filter_word(*args):
    words = []
    for file in args:
        with open(file, 'r', encoding="utf8") as f:
            data = csv.reader(f)
            for row_data in data:
                current_word = ""
                for char in row_data[0]:
                    if char >= 'a' and char <= 'z' or char[0][0] >= 'A' and char[0][0] <= 'Z':
                        current_word += char
                words.append(current_word)

   
    alpha = []
    unique_words = list(set(words))
    unique_words.sort()


    for i in unique_words:
             #yield (i)
        if len(i) > 0:
            alpha.append(i)
                
    print("Total filter records:".format(len(alpha)))
        
    return alpha
```

```python
filter_words = filter_word('vocab.enron.txt')
filter_words
```

### que4 = create a tuple set of all the records avaialble in all the five file and then store it in sqllite DB . (aah,>=,354,fdsf,wer)

```python
import csv

def task4(*args):
    combine_words = []
    for file in args:
        with open(file, 'r', encoding="utf8") as f:
            data = csv.reader(f)
            words = []
            for row_data in data:
                words.append(row_data[0])
            combine_words.append(words)
    

                
    return (combine_words)
```

```python
file = task4('vocab.enron.txt', 'vocab.kos.txt', 'vocab.nips.txt', 'vocab.nytimes.txt', 'vocab.pubmed.txt')
```

```python
f1 = file[0]
f2 = file[1]
f3 = file[2]
f4 = file[3]
f5 = file[4]
```

```python
f1
```

```python
final_zip_list = list(zip(f1, f2, f3, f4, f5))
final_zip_list
```

```python
SQLite Database and Table Creation
Dumping the above data into the Table
```

```python
import sqlite3


db = sqlite3.connect('vocab_database.db')

cursor = db.cursor()
query = "CREATE TABLE new_table(file1 text, file2 text, file3 text, file4 text, file5 text)"
cursor.execute(query)

record_count = 0 
for record in final_zip_list:
    query = "INSERT INTO new_table VALUES {}".format(tuple(record))
    cursor.execute(query)
    record_count += 1
        
db.commit()
    
print('Total records inserted: ' +  str(record_count))
        

db.close()
```
